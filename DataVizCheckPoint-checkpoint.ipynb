```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv('titanic-passengers.csv', delimiter=';')
```

# Preprocessing phase

```python
# Checking if there are missing values
df.isnull().sum().sum()
```

```python
# Have a good look on the dataSet
df.info()
```

```python
# It seems like we have NaN values in the 'Age', 'Cabin' and 'Embarked'  Columns
```

```python
# Let's fill these missing values and drop the Cabin column since we lack a lot of information
df['Age'].fillna(df['Age'].mean(), inplace = True)
df['embarked'] = df.Embarked.fillna('S')
new_df = df.dropna(axis=1, how='any')
new_df.info()
#Description of the Dataset
#survival - Survival (0 = No; 1 = Yes)
#class - Passenger Class (1 = 1st; 2 = 2nd; 3 = 3rd)
#name - Name
#sex - Sex
#age - Age
#sibsp - Number of Siblings/Spouses Aboard
#parch - Number of Parents/Children Aboard
#ticket - Ticket Number
#fare - Passenger Fare
#embarked - Port of Embarkation (C = Cherbourg; Q = Queenstown; S = Southampton)
```

# General Data Visualizations

```python
sns.set(style="darkgrid")
Graph = sns.catplot(x="Pclass",y="Age" ,hue="Sex", col="Survived", data=new_df, height=4, aspect=.7)
```

# def plot_correlation_map

```python
def plot_correlation_map( df ):
    corr = df.corr()
    s , ax = plt.subplots( figsize =( 12 , 10 ) )
    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )
    s = sns.heatmap(
        corr, 
        cmap = cmap,
        square=True, 
        cbar_kws={ 'shrink' : .9 }, 
        ax=ax, 
        annot = True, 
        annot_kws = { 'fontsize' : 12 }
        )
plot_correlation_map( df )
```

# Adding a title column

```python
Title_Dictionary = {

                    "Capt":       "Officer",

                    "Col":        "Officer",

                    "Major":      "Officer",

                      "Dr":         "Officer",

                    "Rev":        "Officer",

                    "Jonkheer":   "Royalty",

                    "Don":        "Royalty",

                    "Sir" :       "Royalty",

                   "Lady" :      "Royalty",

                  "the Countess" : "Royalty",

                    "Dona":       "Royalty",

                    "Mme":        "Miss",

                    "Mlle":       "Miss",

                    "Miss" :      "Miss",

                    "Ms":         "Mrs",

                    "Mr" :        "Mrs",

                    "Mrs" :       "Mrs",

                    "Master" :    "Master"

                    }
```

```python

Titles = []
Title_Dictionary_list = ['Capt', 'Col', 'Major', 'Dr', 'Rev', 'Jonkheer', 'Don', 'Sir', 'Lady', 'the Countess', 'Dona', 'Mme', 'Mlle', 'Miss', 'Ms', 'Mr', 'Mrs', 'Master']
for i in range(len(new_df['Name'])):
    for j in range(len(Title_Dictionary)):
        if  Title_Dictionary_list[j] in new_df['Name'][i]:
            Titles.append(Title_Dictionary[Title_Dictionary_list[j]])
            break
new_df['title'] = Titles   
new_df.info()
```

# Visualize some data using titles

```python
sns.set_style('dark')
surv = new_df[df.Survived == 'Yes']
sns.countplot(x='title', color='white', data=new_df, order=['Officer','Royalty','Miss','Mrs','Master'])
sns.countplot(x='title', data=surv,  order=['Officer','Royalty','Miss','Mrs','Master'])
plt.title('Survival numbers based on Title')
```
